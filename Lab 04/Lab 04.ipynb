```python
from tensorflow.keras.datasets import fashion_mnist
from tsnecuda import TSNE
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
X_train, X_test = X_train.reshape(-1, 784), X_test.reshape(-1, 784)
X, y = np.vstack((X_train, X_test)), np.hstack((y_train, y_test))
label_name = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 5, figsize = (10, 8))
for i, ax in enumerate(axes.flatten()):
    subset = X[y == i][0]
    ax.imshow(subset.reshape((28, 28)), cmap = plt.cm.gray)
    ax.set(xticks = [], yticks = [], title = label_name[i])
plt.tight_layout()
plt.show()
```

```python
start = time.time()
X_embeded = TSNE(n_components = 2, perplexity = 40, n_iter = 5000, learning_rate = 200, random_seed = 0).fit_transform(X)
end = time.time()
print(f'>> Elapsed time = {end - start}s')
```

```python
fig, ax = plt.subplots(figsize = (12, 10))
for label in np.unique(y):
    ax.scatter(X_embeded[y == label][:, 0], X_embeded[y == label][:, 1], label = f'{label_name[label]}', s = 0.5)
# Adjust legend
lgnd = plt.legend(loc = "lower left", scatterpoints = 1, fontsize = 10)
for i in range(len(lgnd.legendHandles)):
    lgnd.legendHandles[i]._sizes = [30]
fig.savefig('tsne-fashion-mnist.jpg')
```

```python
# Try different values of perplexity
embededs = []
perps = list(range(5, 51, 5))
total_elapsed = 0
for perp in perps:
    print(f'>> t-SNE with perplexity = {perp}...', end = '')
    start = time.time()
    X_embeded = TSNE(n_components = 2, perplexity = perp, n_iter = 5000, learning_rate = 200, random_seed = 0).fit_transform(X)
    end = time.time()
    embededs.append(X_embeded)
    print(f'Done!...Elapsed time = {end - start}s')
    total_elapsed += (end - start)
print(f'>> Total elapsed time = {total_elapsed}s')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 5, figsize = (14, 8))
for i, ax in enumerate(axes.flatten()):
    for label in np.unique(y):
        ax.scatter(embededs[i][y == label][:, 0], embededs[i][y == label][:, 1], s = 0.5)
    ax.set(title = f'Perplexity = {perps[i]}', xticks = [], yticks = [])
fig.savefig('tsne-fashion-mnist-different-perp.jpg')
```
