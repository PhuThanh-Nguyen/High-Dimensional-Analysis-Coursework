## b) Using all attributes from the initial dataset except for the Private attribute, do the K-mean clustering method with two clusters. After this step, we will split the original data into two groups (private or not).

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(
    n_clusters = 2, 
    init = 'random', 
    random_state = 0, 
    algorithm = 'full', 
    n_init = 100
).fit(data.iloc[:, 2:])
```

```python
data['Cluster group'] = kmeans.predict(data.iloc[:, 2:])
```

```python
group_0_name = data.loc[data['Cluster group'] == 0, 'University name'].to_list()
group_1_name = data.loc[data['Cluster group'] == 1, 'University name'].to_list()
ivy_league = (
    'Brown University', 
    'Columbia University', 
    'Cornell College', 
    'Dartmouth College',
    'Harvard University',
    'University of Pennsylvania',
    'Princeton University',
    'Yale University'
)
```

```python
for uni in ivy_league:
    if uni in group_0_name:
        print(f'{uni} in Group 1')
    elif uni in group_1_name:
        print(f'{uni} in Group 2')
    else:
        print(f'{uni} not in dataset')
```
